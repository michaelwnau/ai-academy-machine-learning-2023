## MDP from W9 S1 #1


```python
start_state = 3
current_state = start_state
alpha_learning_rate = .5
gamma_discount = .8

Q_s1_right = 0

Q_s2_left = 0
Q_s2_right = 0

Q_s3_left = 0
Q_s3_right = 0

Q_s4_left = 0
Q_s4_right = 0

Q_s5_left = 0
```

```python
def update_Q_s1_right(Q_s1_right):
    return (1-alpha_learning_rate)*Q_s1_right + alpha_learning_rate*(1 + gamma_discount*max(Q_s2_left, Q_s2_right))

def update_Q_s2_left(Q_s2_left):
    return (1-alpha_learning_rate)*Q_s2_left + alpha_learning_rate*(2 + gamma_discount*Q_s1_right)

def update_Q_s2_right(Q_s2_right):
    return (1-alpha_learning_rate)*Q_s2_right + alpha_learning_rate*(1 + gamma_discount*max(Q_s3_left, Q_s3_right))

def update_Q_s3_left(Q_s3_left):
    return (1-alpha_learning_rate)*Q_s3_left + alpha_learning_rate*(2 + gamma_discount*max(Q_s2_left, Q_s2_right))

def update_Q_s3_right(Q_s3_right):
    return (1-alpha_learning_rate)*Q_s3_right + alpha_learning_rate*(2 + gamma_discount*max(Q_s4_left, Q_s4_right))

def update_Q_s4_left(Q_s4_left):
    return (1-alpha_learning_rate)*Q_s4_left + alpha_learning_rate*(3 + gamma_discount*max(Q_s3_left, Q_s3_right))

def update_Q_s4_right(Q_s4_right):
    return (1-alpha_learning_rate)*Q_s4_right + alpha_learning_rate*(20 + gamma_discount*Q_s5_left)

def update_Q_s5_left(Q_s5_left):
    return (1-alpha_learning_rate)*Q_s5_left + alpha_learning_rate*(2 + gamma_discount*max(Q_s4_left, Q_s4_right))

def print_results():
    print('s1 R=', Q_s1_right, '  s2 L=', Q_s2_left, 
          '  s2 R=', Q_s2_right, '  s3 L=', Q_s3_left, 
          '  s3 R=', Q_s3_right, '  s4 L=', Q_s4_left, 
          '  s4 R=', Q_s4_right, '  s5 L=', Q_s5_left, 
          '  Current State = ', current_state)
  
```

```python
Q_s3_left = update_Q_s3_left(Q_s3_left)
current_state -= 1
print_results()

Q_s2_left = update_Q_s2_left(Q_s2_left)
current_state -= 1
print_results()

Q_s1_right = update_Q_s1_right(Q_s1_right)
current_state += 1
print_results()

Q_s2_left = update_Q_s2_left(Q_s2_left)
current_state -= 1
print_results()

Q_s1_right = update_Q_s1_right(Q_s1_right)
current_state += 1
print_results()

Q_s2_left = update_Q_s2_left(Q_s2_left)
current_state -= 1
print_results()

Q_s1_right = update_Q_s1_right(Q_s1_right)
current_state += 1
print_results()
```
